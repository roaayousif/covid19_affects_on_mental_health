```python
#filter for data from april 2020 to april 2023
mental_health_survey = mental_health_survey.loc[mental_health_survey['Time Period'] <= 57]
mental_health_survey
```

```python
# changed 'Time Period Start Date' column to date time data type from oject 
mental_health_survey['Time Period Start Date'] = pd.to_datetime(mental_health_survey['Time Period Start Date'])
```

```python
#changed 'Time Period End Date' column to date time data type from oject 
mental_health_survey['Time Period End Date'] = pd.to_datetime(mental_health_survey['Time Period End Date'])
```

```python
# replacing for consistency. Confirmed these fall under phase 3.0
phase_mapping = {
    '3 (Oct 28 - Dec 21)': '3',
    '3 (Jan 6 - Mar 29)': '3'}
```

```python
mental_health_survey['Phase'] = mental_health_survey['Phase'].replace(phase_mapping)
```

```python
#changing the data type of the phase colun to a float 
mental_health_survey['Phase'] = mental_health_survey['Phase'].astype(float)
```
