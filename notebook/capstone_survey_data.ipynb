```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mental_health_survey = pd.read_csv('../data/Indicators_of_Anxiety_or_Depression_Based_on_Reported_Frequency_of_Symptoms_During_Last_7_Days (4).csv')
```

```python
#filter for data from april 2020 to april 2023
mental_health_survey = mental_health_survey.loc[mental_health_survey['Time Period'] <= 57]
mental_health_survey
```

```python
mental_health_survey['Group'].unique()
```

```python
# dropped columns not needed
mental_health_survey = mental_health_survey[['Indicator', 'Group', 'State', 'Subgroup', 'Phase', 'Time Period',
       'Time Period Label', 'Time Period Start Date', 'Time Period End Date',
       'Value']]
```

```python
mental_health_survey.dtypes
```

```python
# changed 'Time Period Start Date' column to date time data type from oject 
mental_health_survey['Time Period Start Date'] = pd.to_datetime(mental_health_survey['Time Period Start Date'])
```

```python
#changed 'Time Period End Date' column to date time data type from oject 
mental_health_survey['Time Period End Date'] = pd.to_datetime(mental_health_survey['Time Period End Date'])
```

```python
# replacing for consistency. Confirmed these fall under phase 3.0
phase_mapping = {
    '3 (Oct 28 - Dec 21)': '3',
    '3 (Jan 6 - Mar 29)': '3'}
```

```python
mental_health_survey['Phase'] = mental_health_survey['Phase'].replace(phase_mapping)
```

```python
#changing the data type of the phase colun to a float 
mental_health_survey['Phase'] = mental_health_survey['Phase'].astype(float)
```

```python
mental_health_survey.dtypes
```

```python
#looking for missing values 
mental_health_survey.isna().sum()
```

```python
#investigated nans and found that survey was not sent out during those dates ( dates in between phases)
mental_health_survey[mental_health_survey['Value'].isna()]
```

```python
#dropping nans 
mental_health_survey= mental_health_survey.dropna() 
mental_health_survey
```
