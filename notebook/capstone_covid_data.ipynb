```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_df = pd.read_csv('../data/covid_confirmed_usafacts.csv')
```

```python
covid_df
```

```python
covid_df = pd.melt(frame=covid_df,id_vars= ('countyFIPS','County Name', 'State', 'StateFIPS'), var_name= "date", value_name= "cases")
```

```python
covid_df
```

```python
#combined date to get cases number by state instead of by county 
merged_covid_df = covid_df.groupby(['State', 'date'])['cases'].sum().reset_index()
merged_covid_df
```

```python
#filtering for dates 
covid_cases_by_state_df = merged_covid_df.loc[(merged_covid_df['date'] >= '2020-03-01') & (merged_covid_df['date'] <= '2023-04-23')].reset_index(drop=True)
covid_cases_by_state_df
```

```python
covid_cases_by_state_df.dtypes
```

```python
covid_cases_by_state_df['date'] = pd.to_datetime(covid_cases_by_state_df['date'])
```

```python
covid_cases_by_state_df.dtypes
```

```python
covid_cases_by_state_df.to_csv('../data/covid_cases_by_state_df.csv')
```
